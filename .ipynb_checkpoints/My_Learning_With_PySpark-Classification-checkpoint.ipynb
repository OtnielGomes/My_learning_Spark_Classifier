### Defining Functions to Calculate Metrics

#### Confusion Matrix

```python
def confusion_matrix(df_transform_model, normalize=False, percentage=True):
    tp = df_transform_model\
        .select(
            'label',
            'prediction'
        )\
        .where((f.col('label') == 1) & (f.col('prediction') == 1))\
        .count()
    tn = df_transform_model\
        .select(
            'label',
            'prediction'
        )\
        .where((f.col('label') == 0) & (f.col('prediction') == 0))\
        .count()
    fp = df_transform_model\
        .select(
            'label',
            'prediction'
        )\
        .where((f.col('label') == 0) & (f.col('prediction') == 1))\
        .count()
    fn = df_transform_model\
        .select(
            'label',
            'prediction'
        )\
        .where((f.col('label') == 1) & (f.col('prediction') == 0))\
        .count()

    valueP = 1
    valueN = 1
    
    if normalize:
        valueP = tp + fn
        valueN = fp + tn
    
    if percentage and normalize:
        valueP = valueP / 100
        valueN = valueN / 100
    
    print(' '*20, 'Prediction')
    print(' '*15, 'Churn', ' '*5 ,'Not-Churn')
    print(' '*4, 'Churn', ' '*6, int(tp/valueP), ' '*7, int(fn/valueP))
    print('Real')
    print(' '*4, 'Not-Churn', ' '*2, int(fp/valueN), ' '*7, int(tn/valueN))
```

#### Metrics Report

```python
evaluator = MulticlassClassificationEvaluator()
```

```python
def metrics(name_algorithm, predict_training, predict_test):
    print(name_algorithm)
    print("="*40)
    print("Training Data")
    print("="*40)
    print("Confusion Matrix")
    print("-"*40)
    confusion_matrix(predict_training)
    print("-"*40)
    print("Metrics")
    print("-"*40)
    print('Accuracy: %f' % evaluator.evaluate(predict_training, {evaluator.metricName: 'accuracy'}))
    print('Precision: %f' % evaluator.evaluate(predict_training, {evaluator.metricName: 'precisionByLabel', evaluator.metricLabel: 1}))
    print('Recall: %f' % evaluator.evaluate(predict_training, {evaluator.metricName: 'recallByLabel', evaluator.metricLabel: 1}))
    print('F1: %f' % evaluator.evaluate(predict_training, {evaluator.metricName: 'fMeasureByLabel', evaluator.metricLabel: 1}))
    print("")
    print("="*40)
    print("Test Data")
    print("="*40)
    print("Confusion Matrix")
    print("-"*40)
    confusion_matrix(predict_test)
    print("-"*40)
    print("Metrics")
    print("-"*40)
    print('Accuracy: %f' % evaluator.evaluate(predict_test, {evaluator.metricName: 'accuracy'}))
    print('Precision: %f' % evaluator.evaluate(predict_test, {evaluator.metricName: 'precisionByLabel', evaluator.metricLabel: 1}))
    print('Recall: %f' % evaluator.evaluate(predict_test, {evaluator.metricName: 'recallByLabel', evaluator.metricLabel: 1}))
    print('F1: %f' % evaluator.evaluate(predict_test, {evaluator.metricName: 'fMeasureByLabel', evaluator.metricLabel: 1}))
```

### Cross Validation

#### Decision Tree Tuning

```python
dtc_cv = DecisionTreeClassifier(seed=Seed)
```

```python
grid_dtc = ParamGridBuilder()\
    .addGrid(dtc_cv.maxDepth, [2, 5, 10])\
    .addGrid(dtc_cv.maxBins, [10, 32, 45])\
    .build()
```

```python
CV_dtc = CrossValidator(
    estimator=dtc_cv,
    estimatorParamMaps=grid_dtc,
    evaluator=evaluator,
    numFolds=3,
    seed=Seed
)
```

```python
model_dtc_CV = CV_dtc.fit(training)
```

```python
predict_dtc_CV_training = model_dtc_CV.transform(training)
```

```python
predict_dtc_CV_test = model_dtc_CV.transform(test)
```

```python
predict_dtc_CV_training.show(n=10)
```

```python
predict_dtc_CV_test.show(n=10)
```

```python
metrics('Decision Tree Tuning', predict_dtc_CV_training, predict_dtc_CV_test)
```

#### Random Forest Tuning

```python
rfc_cv = RandomForestClassifier(seed=Seed)
```

```python
grid_rfc = ParamGridBuilder()\
    .addGrid(rfc_cv.numTrees, [10, 20, 40, 100, 500])\
    .addGrid(rfc_cv.maxDepth, [2, 5, 10])\
    .addGrid(rfc_cv.maxBins, [10, 32, 45])\
    .build()
```

```python
CV_rfc = CrossValidator(
    estimator=rfc_cv,
    estimatorParamMaps=grid_rfc,
    evaluator=evaluator,
    numFolds=3,
    seed=Seed
)
```

```python
model_rfc_CV = CV_rfc.fit(training)
```

```python
predict_rfc_CV_training = model_rfc_CV.transform(training)
```

```python
predict_rfc_CV_test= model_rfc_CV.transform(test)
```

```python
predict_rfc_CV_training.show(n=10)
```

```python
predict_rfc_CV_test.show(n=10)
```

```python
metrics('Random Forest Tuning', predict_rfc_CV_training, predict_rfc_CV_test)
```
